```python
#Prahlad Pandav, ME 592, HW1
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2
from google.colab.patches import cv2_imshow

#Problem 1
img = cv2.imread('sudoku-original.png') #read image file
g = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #convert original image to grayscale
print('Original Image:')
cv2_imshow(img) #show original image
print('Grayscale Image:')
cv2_imshow(g) #show grayscale image
cv2.waitKey(0)
cv2.destroyAllWindows()

#Histogram of Intensities
h = plt.hist(img.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k') #defining histogram
c = cv2.calcHist([img], [0], None, [256], [0,256]) #calculates histogram parameters
plt.plot(c) #plot histogram
plt.title('Intensity Histogram') #title of histogram
plt.xlabel('Frequency, Hz') #title of x-axis
plt.ylabel('Pixel Count') #title of y-axis
plt.show() #display histogram

#Perspective Transformation
p1 = np.float32([[55, 65], [367, 52], [25, 388], [390, 390]]) #Pixel coordinates of image obtained from MS Paint
p2 = np.float32([[0, 0], [419, 0], [0, 423], [419, 423]]) #Desired pixel coordinates for perspective transformss

m = cv2.getPerspectiveTransform(p1, p2) #Perspective transforms image
f = cv2.warpPerspective(img, m, (419, 423)) #Apply perspective transformation to image
print('Original Image:')
cv2_imshow(img)
print('Perspective Transformed Image:')
cv2_imshow(f)
cv2.waitKey(0)
```

```python
#Part A: Find Mean
dat = np.loadtxt('airfoil_self_noise.dat', unpack=True) #load and read .dat file
mean = dat.mean() #calculates mean
print('The mean is: ', mean)

#Part B: Find Standard Deviation
sig = dat.std() #calculates standard deviation
print('The standard deviation is: ', sig)

#Part C: Find Median
med = np.median(dat) #calculates median
print('The median is: ', med)

#Part D: Find Kurtosis
dataFrame = pd.DataFrame(dat)
k = dataFrame.kurt(axis=1) #calculates kurtosis
print('Kurtosis: ', k)

#Part E: Find Skewness
sk = pd.DataFrame(data=dat)
skewValue = dataFrame.skew(axis=1) #finds skewness of data
print('The skewness is: ', skewValue)
```
